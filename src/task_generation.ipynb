# Module importing

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
```

```python
%matplotlib inline
```

# Loading linear data

```python
df = pd.read_csv('data/restaurant_revenue.txt', sep=',', header=None)
df.columns = ['x', 'y']
```

```python
lr = LinearRegression()
model = lr.fit(df[['x']].values, df['y'].values)
```

```python
preds = model.predict(np.linspace(0, 23, 50).reshape(-1, 1))
```

```python
model.intercept_, model.coef_
```

```python
fig, ax = plt.subplots(1, 1, figsize = (15, 8))
ax.scatter(df['x'], df['y'], c='green', label='точка ($x_i$, $y_i$)')
ax.plot(np.linspace(0, 23, 50), preds, c='red', label='h(x) = $-3.9*x + 1.2$')
plt.title('Апроксимація точок лінійною функцією')
ax.legend()
plt.savefig('img/linear.jpg')
```

# Generation of quadratic dependency

```python
# create 1000 equally spaced points between 0 and 50
X = np.linspace(0, 50, 100)
```

```python
# calculate the y value for each element of the x vector
y = (-1)*X**2 + 100*X + 50
```

```python
x_noize = X + np.random.normal(0, 5, size=X.size)
y_noize = y + np.random.normal(0, 3, size=y.size)
```

```python
data_noize = pd.DataFrame({'x': x_noize, 'y': y_noize})
```

```python
fig, ax = plt.subplots(1, 1, figsize = (15, 8))
ax.scatter(x_noize, y_noize, c='green', label='точка ($x_i$, $y_i$)')
ax.plot(X, y, c='red', label='h(x) = $-x^2 + 100*x + 50$')
plt.title('Апроксимація точок квадратичною функцією')
ax.legend()
plt.savefig('img/quadratic.jpg')
```
